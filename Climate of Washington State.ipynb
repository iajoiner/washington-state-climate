# 1. Process the site list to obtain a list of webpages to download

```python
from bs4 import BeautifulSoup
with open("site_list.html", "r") as f:
    html_doc = f.read()
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
station_list = soup.find_all('a')
```

```python
name_list = []
link_list = []
for station in station_list:
    name_list.append(station.contents[0].strip())
    link_list.append('https://wrcc.dri.edu/cgi-bin/cliRECtM.pl?' + station['href'][-6:])
```

```python
del name_list[-1]
del link_list[-1]
```

```python
print(len(name_list))
print(len(link_list))
```

# 2. Download and process the pages

```python
import requests
```

```python
num = len(name_list)
for i in range(num):
    csv_string = ',Jan,Feb,Mar,Apr,May,Jun,Jul,Aug,Sep,Oct,Nov,Dec,Total\n'
    r = requests.get(link_list[i])
    print('No.' + str(i) + ' ' + name_list[i] + ' is getting processed!')
    html_file = r.text
    soup2 = BeautifulSoup(html_file, 'html.parser')
    if (link_list[i][-6] == 'i'):
        file_name = 'data/ID ' + name_list[i] + '.csv'
    elif (link_list[i][-6] == 'o'):
        file_name = 'data/OR ' + name_list[i] + '.csv'
    else:
        file_name = 'data/WA ' + name_list[i] + '.csv'
    with open(file_name, "w+") as f:
        table_rows = soup2.find_all('tr')
        del table_rows[0]
        for row in table_rows:
            row_content = str(row.contents[1])
            processed_content = row_content.replace('<td>','').replace('<td align="RIGHT">','').replace('</td>','')
            #print(processed_content)
            list_content = processed_content.split('\n')
            new_list = list(map(str.strip, list_content))
            del new_list[-1]
            csv_row = ','.join(new_list)
            csv_string = csv_string + csv_row + '\n'
        f.write(csv_string)            
```

```python
with open("station_list.csv", "w+") as g:
    name_string = ','.join(name_list)
    g.write(name_string)
```

Now it is time to extract data about the stations in order to form a shared data frame.

```python
import pandas as pd
```

```python
length = len(link_list)
for i in range(length):
    link_list[i] = link_list[i][-6:]
df_weather = pd.DataFrame({'Station': name_list, 'ID': link_list})
```

```python
df_weather
```

```python
for i in range(length):
    df_weather.at[i,'State'] = df_weather.at[i,'ID'][:2].upper()
    df_weather.at[i,'File'] = df_weather.at[i,'State'] + ' ' + df_weather.at[i,'Station']
```

```python
def loc_extractor(ID):
    url = 'https://wrcc.dri.edu/cgi-bin/cliMeta.pl?' + ID
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find('table')
    rows = table.find_all('tr')
    last_row = rows[-1]
    entries = last_row.find_all('td')
    lat = int(entries[3].string)/100
    long = -int(entries[4].string)/100
    return lat, long
```

```python
for i in range(length):
    print(i)
    print(df_weather.loc[i, 'Station'])
    df_weather.loc[i, 'Latitude'], df_weather.loc[i, 'Longitude'] = loc_extractor(df_weather.loc[i, 'ID'])
```

```python
df_weather.to_csv('stations.csv')
```

# 3. Graphing

We will use Folium to plot the graphs.

```python
import folium
mapping = folium.Map(location = [45.6272, -119.1199], zoom_start = 6)
```

```python
def gen_label(pos):
    with open('html/' + df_weather.loc[pos, 'File'] + '.html', 'r') as f:
        html = f.read()
    return df_weather.loc[pos, 'Station'] + '\n' + html
```

```python
def csv2html(pos):
    print(pos)
    print(df_weather.loc[pos, 'Station'])
    df_temp = pd.read_csv('data/' + df_weather.loc[pos, 'File'] + '.csv', index_col = 0)
    df_temp.to_html('html/' + df_weather.loc[pos, 'File'] + '.html')
```

```python
for i in range(length):
    csv2html(i)
    folium.CircleMarker(
    location=[df_weather.loc[i, 'Latitude'], df_weather.loc[i, 'Longitude']],
    radius=1,
    popup=gen_label(i),
    color='#3186cc',
    fill=True,
    fill_color = '#3186cc'
).add_to(mapping)
```
